### Text Pre-Processing
##### Stop-words removal 

```python
stopwords = open('stopwords_en.txt','r').read().split('\n')
stopwords
```

```python
ss = data.iloc[0,0].split()
ss
```

```python
filtered_words = []
for s in ss:
    if s in stopwords:
        pass
    else:
        filtered_words.append(s)
```

```python
filtered_words
```

```python
def remove_stop_words(text):
    filtered_words = []
    words = text.split()
    for word in words:
        if word.lower() in stopwords:
            pass
        else:
            filtered_words.append(word)
    return ' '.join(filtered_words)
```

```python
remove_stop_words(data.iloc[0,0])
```

```python
data['text_no_sw'] = data.Text.apply(remove_stop_words)
data.head()
```

##### Stemming

```python
from nltk.stem import PorterStemmer
```

```python
sm = PorterStemmer()
```

```python
sm.stem("ezayak")
```

```python
def cleaning(text):
    text  =  re.sub('[^a-zA-Z\']',' ',text).replace("  ",' ')
    filtered_words = []
    words = text.split()
    for word in words:
        if word.lower() in stopwords:
            pass
        else:
            filtered_words.append(sm.stem(word))
    return ' '.join(filtered_words)
```

```python
cleaning(data.iloc[1,0])
```

```python
data['clean_text'] = data.Text.apply(cleaning)
data.head()
```

##### Arabic

```python
from nltk.stem import ISRIStemmer
```

```python
stopwords = open("stopwords_ar.txt",'r').read().split('\n') ## 1
sm_ar = ISRIStemmer() ## 2
def cleaning(text):
    text  =  re.sub('[^ء-ي]',' ',text).replace("  ",' ')
    words = text.split()
    for word in words:
        if word in stopwords:
            pass
        else:
            filtered_words.append(sm_ar.stem(word))
    return ' '.join(filtered_words)
```

```python
stopwords
```

```python
data_ar = pd.read_csv('dataset.csv',names=['text','label'])
data_ar.head()
```

```python
data_ar['clean_text'] = data_ar['text'].apply(cleaning)
data_ar.head()
```

```python
data_ar[data_ar["clean_text"]==""]
```

```python
data_ar.drop(data_ar[data_ar["clean_text"]==""].index,inplace=True)
```

```python
data_ar[data_ar["clean_text"]==""]
```

```python
x = tfidf.fit_transform(data_ar.clean_text).toarray()
x.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,data_ar.label,test_size = 0.2,random_state=30)
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```
