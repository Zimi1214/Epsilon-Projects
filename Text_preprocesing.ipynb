## Why preprocess?
- Helps make for better input data
- When performing machine learning or other statistical methods
Examples:
Tokenization to create a bag of words
- Lowercasing words
- Lemmatization/Stemming
    - Shorten words to their root stems
- Removing stop words, punctuation, or unwanted tokens
- Good to experiment with different approaches

# Stemming and Lemmatization

## "Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the Language."
- (-ed,-ize, -s,-de,mis)

## Lemmatization, unlike Stemming, reduces the inflected words properly ensuring that the root word belongs to the language.

```python
# Import WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
```

```python
# Retain alphabetic words: alpha_only
alpha_only = [t for t in lower_tokens if t.isalpha()]
alpha_only
```

```python
#load NLTK stop words : english_stops
from nltk.corpus import stopwords
english_stops = set(stopwords.words('english'))
```

```python
english_stops
```

```python
# Remove all stop words: no_stops
no_stops = [t for t in alpha_only if t not in english_stops]
```

```python
no_stops
```

```python
Counter(no_stops)
```

```python
from nltk.stem import PorterStemmer

porter=PorterStemmer()

stem = [porter.stem(t) for t in no_stops]

print(no_stops[:200])
print("*"*50)
print(stem[:200])
```

```python
# Instantiate the WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# Lemmatize all tokens into a new list: lemmatized
lemmatized = [wordnet_lemmatizer.lemmatize(t) for t in no_stops]

print(no_stops[:200])
print("*"*50)
print(lemmatized[:200])
```

```python
import matplotlib
from nltk.probability import FreqDist
matplotlib.use('TkAgg') 

def plot_freq_dist(words, num_words = 20):
    fdist = FreqDist(words)
    fdist.plot(num_words,cumulative=False)
```

```python
plot_freq_dist(lemmatized, num_words=30)
```
